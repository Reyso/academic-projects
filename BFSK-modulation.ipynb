```python
# Importando as bibliotecas
import matplotlib.pyplot as plt # plotar imagens
import numpy as np # biblioteca matematica
from math import pi #pi
```

```python
# Parâmetros do gráfico

plt.figure(figsize=(15,10)) # tamanho da figura
plt.style.use('fivethirtyeight') # estilo



# vetor de sinal digital
v = np.array([1, 0, 1, 1, 0, 1, 0, 0])

# Número de amostras o Sinal
N = len(v)



# Amostras do pulso biário
dim=1000


#vetores de suporte para o código
Vx=[]
Di=[]

# Frequência (hz)
  ## Frequência Portadora 1
f1= 3
  ## Frequência Portadora 2
f2= 10

# Sinal Digital
for i in range (1,len(v)):
    f=np.ones(dim)
    x=f*v[i]
    Vx=np.concatenate((Vx,x))

# eixo x
dim2=len(Vx)
t=np.linspace(0,5,dim2)

# Sinal Portadora 1
  ## Forma do sinal
w1=2*np.pi*f1*t
y1=np.sin(w1)

# Sinal Portadora 2
  ## Forma do sinal
w2=2*np.pi*f2*t 
y2=np.sin(w2)



# Sinal de Saída
plt.subplot(4,1,4)
for i in range(0,dim2):
    if Vx[i]==0:
        zero=np.array([y1[i]])
        Di=np.concatenate((Di,zero))
    else:
        one=np.array([y2[i]])
        Di=np.concatenate((Di,one))


# Plot Sinal Digital 
plt.subplot(4,1,1)
plt.plot(Vx)
plt.ylabel('S_Digital')

# Plot portadora 1
plt.subplot(4,1,2)
plt.plot(t,y1)
plt.ylabel('S_Portadora 1')

# Plot portadora 2
plt.subplot(4,1,3)
plt.plot(t,y2)
plt.ylabel('S_Portadora 2')


# plot sinal modulado
plt.subplot(4,1,4)
plt.plot(t,Di)
plt.plot(t,Vx)
plt.ylabel('S_Modulado-BFSK')


plt.grid(True)
plt.show()

```

```python
#
plt.figure(figsize=(10,3)) # tamanho da figura
# Sinal digital
plt.xlabel('Tempo(s)')
plt.ylabel('Amplitude')
plt.title('Sinal Digital - Informação a ser enviada')
plt.plot(Vx)
```

```python
# Portadoras modulantes
plt.figure(figsize=(10,3))
plt.xlabel('Tempo(s)')
plt.ylabel('Amplitude')
plt.title('Portadora modulante')
plt.plot(t,y1)
```

```python
# Portadoras modulantes
plt.figure(figsize=(10,3))
plt.xlabel('Tempo(s)')
plt.ylabel('Amplitude')
plt.title('Portadora modulante 2 ')
plt.plot(t,y2)
```
